## Generated 100 sequences of 1,000 states based on PI and  A
### Markov model with three states, S1, S2, S3, initial probabilities 
#### PI= [0.5,0.2,0.3] and  A= [0.4   0.3   0.3          0.2    0.6   0.2           0.1    0.1   0.8]

```python
import numpy as np
import random as rm
```

```python
# The statespace
states = ["S1","S2","S3"]

# initial state probabilities
PI= [0.5, 0.2, 0.3]

# Possible sequences of events based on states
transitionName = [["S1S1","S1S2","S1S3"],["S2S2","S2S1","S2S3"],["S3S1","S3S3","S3S2"]]

# Transition matrix
A = [[0.4,0.3,0.3],[0.2,0.6,0.2],[0.1,0.1,0.8]]
```

```python
if sum(A[0])+sum(A[1])+sum(A[1]) != 3:
    print("Probablity Sum Transition Matrix went wrong!")
else: print("Hidden Markov Model!!!!!")
```

## Generated Data based on PI and A

```python
def state_list(st):
    # Choose the starting state
    initial_state = np.random.choice(states,replace=True,p=PI)
    #print(initial_state)
    
    start_List = [initial_state]
    i = 0
    prob = 1
    while i != st:
        if initial_state == "S1":
            change = np.random.choice(transitionName[0],replace=True,p=A[0])
            if change == "S1S1":
                prob = prob * 0.4
                start_List.append("S1")
                pass
            elif change == "S1S2":
                prob = prob * 0.3
                initial_state = "S2"
                start_List.append("S2")
            else:
                prob = prob * 0.3
                initial_state = "S3"
                start_List.append("S3")
        elif initial_state == "S2":
            change = np.random.choice(transitionName[1],replace=True,p=A[1])
            if change == "S2S2":
                prob = prob * 0.2
                start_List.append("S2")
                pass
            elif change == "S2S1":
                prob = prob * 0.6
                initial_state = "S1"
                start_List.append("S1")
            else:
                prob = prob * 0.2
                initial_state = "S3"
                start_List.append("S3")
        elif initial_state == "S3":
            change = np.random.choice(transitionName[2],replace=True,p=A[2])
            if change == "S3S3":
                prob = prob * 0.1
                start_List.append("S3")
                pass
            elif change == "S3S1":
                prob = prob * 0.1
                initial_state = "S1"
                start_List.append("S1")
            else:
                prob = prob * 0.8
                initial_state = "S2"
                start_List.append("S2")
        i += 1    
    return start_List

# To save every state list
list_states = []
count = 0

# Starts range from the first count up until but excluding the last count
for iterations in range(0,100):
        list_states.append(state_list(999))

```

# Building HMM and generating samples By using HMMLearn library 

## Importing HMMLearn library 

```python
import numpy as np
import matplotlib.pyplot as plt
from hmmlearn import hmm
```

## Question Answer 1.a: Generated 100 sequences of 1,000 states based on PI and  A by using HMMLearn library functions


```python
startprob_ = np.array([0.5, 0.2, 0.3])
# The transition matrix, note that there are no transitions possible
# between component 1 and 3
transmat_ = np.array([[0.4, 0.3, 0.3], 
                    [0.2, 0.6, 0.2], 
                    [0.1, 0.1, 0.8]])


# The means of each component
means = np.array([[0.0, 0.0], [3.0, -3.0], [5.0, 10.0]])
# The covariance of each component
covars = np.tile(np.identity(2), (3, 1, 1))

# Build an HMM instance and set parameters
model = hmm.GaussianHMM(n_components=3,n_iter=1000,covariance_type="full")

# Instead of fitting it from the data, we directly set the estimated
# parameters, the means and covariance of the components
model.startprob_ = startprob_
model.transmat_ = transmat_
model.means_ = means
model.covars_ = covars
```

```python
# Generate samples
X, Z = model.sample(100)
```

## Fitting HMMLearn library for sample 100 sequences of 1,000 states.

```python
remodel = hmm.GaussianHMM(n_components=3, covariance_type="full", n_iter=1000)
remodel.fit(X)
```
